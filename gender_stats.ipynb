# Education Indicators vs GDP

```python
#find correlation between "the ratio of female to male labor force participation rate (%) (modeled ILO estimate)" and GDP per capita
correlation = st.pearsonr(f2m_laborforce, gdp_percapita_2020)
print(f"The correlation between female to male labor force participation rate (%) (modeled ILO estimate) and GDP per capita is {round(correlation[0],2)}")
```

```python
#testing the above
np.corrcoef(f2m_laborforce, gdp_percapita_2020)[0][1]
```

```python
#Create a scatter plot to show the realtionship between "the ratio of female to male labor force participation rate (%) (modeled ILO estimate)" and GDP per capita
from scipy.stats import linregress

fig, ax = plt.subplots(figsize=(15, 15))

plt.scatter(f2m_laborforce, gdp_percapita_2020, label='Ratio of female to male labor force participation rate (%)', color='blue')
plt.xlabel('Female to Male Labor Force Participation Ratio')
plt.ylabel('GDP Per Capita')
plt.title('Ratio of female to male labor force participation rate vs GDP Per Capita')

plt.show

#Determine the linear regression
slope, intercept, r_value, p_value, std_err = st.linregress(f2m_laborforce, gdp_percapita_2020)

line_eq = slope * f2m_laborforce + intercept
plt.plot(f2m_laborforce,line_eq, color='red')

print(f"The r-squared is: {r_value**2}")
```

```python
income_data.columns
```

```python
#drop "source" column
income_data_clean = income_data.drop('Source', axis=1)
```

```python
#remove "total" rows
rows_to_drop = income_data_clean[income_data_clean['Sex'].str.contains("Total")].index
income_data_clean.drop(rows_to_drop, inplace=True)
income_data_clean.head(10)
```

```python
income_data_clean.dropna(inplace=True)
income_data_clean[["U.S. dollars"]].apply(pd.to_numeric)
```

```python
#create dataframe with country, male income, female income
gender_list=income_data_clean['Sex'].unique()
country_list = income_data_clean['Reference area'].unique()
d1=[]
d2=[]
d1=income_data_clean.loc[income_data_clean["Sex"]==gender_list[0],'U.S. dollars'].tolist()
d2=income_data_clean.loc[income_data_clean["Sex"]==gender_list[1],'U.S. dollars'].tolist()
genderinc={"Country Name":country_list,"Income Male":d1, "Income Female":d2}
genderinc=pd.DataFrame(data=genderinc)
genderinc
```

```python
#Create a column to take the income ratios and then add to dataframe
ratios=[]
ratios = (genderinc["Income Female"]/genderinc["Income Male"]).tolist()
genderinc['FemaleToMaleIncomeRatio'] = ratios
genderinc
```

```python
merged_data = pd.merge(genderinc,gender_analysis_df, on='Country Name', how='left' )
merged_data
```

```python
merged_data.dropna(inplace=True)
merged_data
```

```python
countryname = merged_data['Country Name']
incomemale = merged_data['Income Male']
incomefemale = merged_data['Income Female']
femaletomale = merged_data['FemaleToMaleIncomeRatio']
gdp = round(merged_data['2020'], 2)
f2m_lbforce = merged_data['Ratio of female to male labor force participation rate (%)']
paygap = ({
    'Country Name' : countryname,
    'Income Male' : incomemale,
    'Income Female' : incomefemale,
    'FemaleToMaleIncomeRatio' : femaletomale,
    '2020 GDP Per Capita (USD)' : gdp
})

paygap_df = pd.DataFrame(paygap)
paygap_df
```

```python
#visualize female vs male income by country using a double bar graph
paygap_df.plot(kind='bar', x='Country Name', y=['Income Female', 'Income Male'], 
        stacked=False, color=['red', 'blue'])
plt.xlabel("Country Name")
plt.ylabel("Income")
plt.title("Monthly Income by Female and Male")
plt.legend()
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for female-to-male income ratio and gdp per capita
correlation = st.pearsonr(femaletomale,gdp)
print(f"The correlation between female-to-male income ratio and gdp per capita is {round(correlation[0],2)}")
```

```python
#visualize the relationship between female-to-male income ratio and gdp per capita using a line graph
paygap_df.plot(kind='line', x='FemaleToMaleIncomeRatio', y='2020 GDP Per Capita (USD)', color='blue')
plt.xlabel("Female-to-Male Income Ratio")
plt.ylabel("2020 GDP Per Capita (USD)")
plt.title("Female-to-Male Income Ratio vs 2020 GDP Per Capita")
plt.legend()
plt.show()
```

```python
#Calculate the linear regression model for female-to-male income ratio and gdp per capita
slope, intercept, r_value, p_value, std_err = st.linregress(femaletomale, gdp)
line_eq = slope * femaletomale + intercept

#visualize the relationship between female-to-male income ratio and gdp per capita using a scatter plot
paygap_df.plot(kind='scatter', x='FemaleToMaleIncomeRatio', y='2020 GDP Per Capita (USD)', color='blue')
plt.xlabel("Female-to-Male Income Ratio")
plt.ylabel("2020 GDP Per Capita (USD)")
plt.title("Female-to-Male Income Ratio vs 2020 GDP Per Capita")
plt.plot(femaletomale,line_eq, color='red')

plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for female-to_male labor force participation vs female-to-male income ratio
correlation = st.pearsonr(f2m_lbforce,femaletomale)
print(f"The correlation between  female-to_male labor force participation vs female-to-male income ratio is {round(correlation[0],2)}")
```

```python
#Calculate the linear regression model for female-to_male labor force participation vs female-to-male income ratio
slope, intercept, r_value, p_value, std_err = st.linregress(f2m_lbforce, femaletomale)
line_eq = slope * f2m_lbforce + intercept

#visualize female-to_male labor force participation vs female-to-male income ratio using a scatter plot
merged_data.plot(kind='scatter', x='Ratio of female to male labor force participation rate (%)', y='FemaleToMaleIncomeRatio', color='blue')
plt.xlabel("Ratio of Female-to-Male Labor Force Participation Rate (%)")
plt.ylabel("Female-to-Male Income Ratio")
plt.title("Female-to-Male Labor Force Participation Rate vs Female-to-Male Income Ratio")
plt.plot(f2m_lbforce,line_eq, color='red')
plt.show()

print(f"The r-squared is: {r_value**2}")
```
